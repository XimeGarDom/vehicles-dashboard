```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv('vehicles_us.csv')

# Vista general
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['odometer'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Histograma del Od√≥metro')
plt.xlabel('Kilometraje')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['odometer'], df['price'], alpha=0.3, color='green')
plt.title('Kilometraje vs Precio')
plt.xlabel('Od√≥metro')
plt.ylabel('Precio')
plt.grid(True)
plt.show()
```
